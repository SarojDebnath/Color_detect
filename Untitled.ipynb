```python
import cv2
import numpy as np

def detect_led_color(image, x, y, radius):
    # Extract the region of interest (ROI) around the circle
    roi_size = int(radius * 2)  # Use the circle's diameter as ROI size
    half_size = roi_size // 2
    # Ensure ROI stays within image bounds
    y_start = max(0, int(y - half_size))
    y_end = min(image.shape[0], int(y + half_size))
    x_start = max(0, int(x - half_size))
    x_end = min(image.shape[1], int(x + half_size))
    roi = image[y_start:y_end, x_start:x_end]

    # Convert ROI to HSV
    hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

    # Define color ranges in HSV (tuned for better red and orange detection)
    red_lower1 = np.array([0, 120, 70])    # Lower red range
    red_upper1 = np.array([10, 255, 255])
    red_lower2 = np.array([170, 120, 70])  # Upper red range
    red_upper2 = np.array([180, 255, 255])
    green_lower = np.array([35, 100, 100])
    green_upper = np.array([85, 255, 255])
    blue_lower = np.array([100, 100, 100])
    blue_upper = np.array([130, 255, 255])
    orange_lower = np.array([5, 120, 70])   # Widened orange range
    orange_upper = np.array([25, 255, 255])

    # Create color masks
    red_mask1 = cv2.inRange(hsv, red_lower1, red_upper1)
    red_mask2 = cv2.inRange(hsv, red_lower2, red_upper2)
    red_mask = cv2.bitwise_or(red_mask1, red_mask2)
    green_mask = cv2.inRange(hsv, green_lower, green_upper)
    blue_mask = cv2.inRange(hsv, blue_lower, blue_upper)
    orange_mask = cv2.inRange(hsv, orange_lower, orange_upper)

    # Calculate the percentage of each color
    red_percent = (red_mask > 0).mean() * 100
    green_percent = (green_mask > 0).mean() * 100
    blue_percent = (blue_mask > 0).mean() * 100
    orange_percent = (orange_mask > 0).mean() * 100

    # Determine the dominant color
    colors = {
        'red': red_percent,
        'green': green_percent,
        'blue': blue_percent,
        'orange': orange_percent
    }
    dominant_color = max(colors, key=colors.get)

    # Define BGR colors for drawing
    color_map = {
        'red': (0, 0, 255),
        'green': (0, 255, 0),
        'blue': (255, 0, 0),
        'orange': (0, 165, 255)
    }

    return dominant_color, colors, color_map[dominant_color]

# Main processing loop
for k in range(6):
    path = f'test/{k}.png'
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    original_image = cv2.imread(path)
    
    if image is None or original_image is None:
        print(f"Failed to load image {path}")
        continue

    # Convert grayscale image to binary using a threshold
    _, binary_image = cv2.threshold(image, 40, 255, cv2.THRESH_BINARY)

    # Morphological operations
    kernel = np.ones((3, 3), np.uint8)
    dilated = cv2.dilate(binary_image, kernel, iterations=1)
    eroded = cv2.erode(dilated, kernel, iterations=1)
    dilated = cv2.dilate(eroded, kernel, iterations=1)

    # Apply Hough Circle Transform
    circles = cv2.HoughCircles(
        dilated,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=5,
        param1=10,
        param2=9,
        minRadius=5,
        maxRadius=10
    )

    # Create a color image for drawing circles
    output = original_image.copy()

    # Process detected circles
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i, circle in enumerate(circles[0, :]):
            center = (circle[0], circle[1])
            radius = circle[2]
            
            # Detect color of the LED
            dominant_color, color_percentages, draw_color = detect_led_color(original_image, center[0], center[1], radius)
            
            # Draw circles with detected color
            cv2.circle(output, center, radius, draw_color, 2)  # Outline in detected color
            cv2.circle(output, center, 1, (0, 255, 0), 2)    # Green center
            # Add color label
            #cv2.putText(output,f"{dominant_color} ({color_percentages[dominant_color]:.1f}%)",(center[0] + radius + 5, center[1]),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255, 255, 255),1,cv2.LINE_AA)

            # Detailed debug output
            print(f"LED {i+1} in image {k} (center: {center}, radius: {radius}):")
            print(f"Dominant color: {dominant_color}")
            print(f"Color percentages: Red: {color_percentages['red']:.1f}%, "
                  f"Green: {color_percentages['green']:.1f}%, "
                  f"Blue: {color_percentages['blue']:.1f}%, "
                  f"Orange: {color_percentages['orange']:.1f}%")

    # Display images
    cv2.imshow('Binary Image', binary_image)
    cv2.imshow('Eroded Image', eroded)
    cv2.imshow('Dilated Image', dilated)
    cv2.imshow(f'Circles with Colors {k}', output)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def detect_led_color(image, x, y, radius):
    # Extract the region of interest (ROI) around the circle
    roi_size = int(radius * 2)  # Use the circle's diameter as ROI size
    half_size = roi_size // 2
    # Ensure ROI stays within image bounds
    y_start = max(0, int(y - half_size))
    y_end = min(image.shape[0], int(y + half_size))
    x_start = max(0, int(x - half_size))
    x_end = min(image.shape[1], int(x + half_size))
    roi = image[y_start:y_end, x_start:x_end]

    # Convert ROI to HSV
    hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

    # Define color ranges in HSV
    red_lower1 = np.array([0, 120, 70])
    red_upper1 = np.array([10, 255, 255])
    red_lower2 = np.array([170, 120, 70])
    red_upper2 = np.array([180, 255, 255])
    green_lower = np.array([35, 100, 100])
    green_upper = np.array([85, 255, 255])
    blue_lower = np.array([100, 100, 100])
    blue_upper = np.array([130, 255, 255])

    # Updated orange range to avoid overlap with red and refine detection
    orange_lower = np.array([11, 150, 150])
    orange_upper = np.array([25, 255, 255])

    # Create color masks
    red_mask1 = cv2.inRange(hsv, red_lower1, red_upper1)
    red_mask2 = cv2.inRange(hsv, red_lower2, red_upper2)
    red_mask = cv2.bitwise_or(red_mask1, red_mask2)
    green_mask = cv2.inRange(hsv, green_lower, green_upper)
    blue_mask = cv2.inRange(hsv, blue_lower, blue_upper)

    # Refined orange mask using saturation and value constraints
    orange_mask_raw = cv2.inRange(hsv, orange_lower, orange_upper)
    saturation = hsv[:, :, 1]
    value = hsv[:, :, 2]
    orange_mask = cv2.bitwise_and(
        orange_mask_raw,
        orange_mask_raw,
        mask=cv2.inRange(saturation, 150, 255) & cv2.inRange(value, 150, 255)
    )

    # Calculate the percentage of each color
    red_percent = (red_mask > 0).mean() * 100
    green_percent = (green_mask > 0).mean() * 100
    blue_percent = (blue_mask > 0).mean() * 100
    orange_percent = (orange_mask > 0).mean() * 100

    # Determine the dominant color
    colors = {
        'red': red_percent,
        'green': green_percent,
        'blue': blue_percent,
        'orange': orange_percent
    }
    dominant_color = max(colors, key=colors.get)

    # Define BGR colors for drawing
    color_map = {
        'red': (0, 0, 255),
        'green': (0, 255, 0),
        'blue': (255, 0, 0),
        'orange': (0, 165, 255)
    }

    return dominant_color, colors, color_map[dominant_color]

# Main processing loop
for k in range(6):
    path = f'test/{k}.png'
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    original_image = cv2.imread(path)
    
    if image is None or original_image is None:
        print(f"Failed to load image {path}")
        continue

    # Convert grayscale image to binary using a threshold
    _, binary_image = cv2.threshold(image, 40, 255, cv2.THRESH_BINARY)

    # Morphological operations
    kernel = np.ones((3, 3), np.uint8)
    dilated = cv2.dilate(binary_image, kernel, iterations=1)
    eroded = cv2.erode(dilated, kernel, iterations=1)
    dilated = cv2.dilate(eroded, kernel, iterations=1)

    # Apply Hough Circle Transform
    circles = cv2.HoughCircles(
        dilated,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=5,
        param1=10,
        param2=9,
        minRadius=5,
        maxRadius=10
    )

    # Create a color image for drawing circles
    output = original_image.copy()

    # Process detected circles
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i, circle in enumerate(circles[0, :]):
            center = (circle[0], circle[1])
            radius = circle[2]
            
            # Detect color of the LED
            dominant_color, color_percentages, draw_color = detect_led_color(original_image, center[0], center[1], radius)
            
            # Draw circles with detected color
            cv2.circle(output, center, radius, draw_color, 2)  # Outline in detected color
            cv2.circle(output, center, 1, (0, 255, 0), 2)    # Green center

            # Debug output
            print(f"LED {i+1} in image {k} (center: {center}, radius: {radius}):")
            print(f"Dominant color: {dominant_color}")
            print(f"Color percentages: Red: {color_percentages['red']:.1f}%, "
                  f"Green: {color_percentages['green']:.1f}%, "
                  f"Blue: {color_percentages['blue']:.1f}%, "
                  f"Orange: {color_percentages['orange']:.1f}%")

    # Display images
    cv2.imshow('Binary Image', binary_image)
    cv2.imshow('Eroded Image', eroded)
    cv2.imshow('Dilated Image', dilated)
    cv2.imshow(f'Circles with Colors {k}', output)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
